```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")

plt.style.use('seaborn-bright')
```

```python
tips.sample(10)
```

```python
plt.figure(figsize=(6,4))
sns.displot(tips['total_bill'],kde=True)
plt.show()
```

joining two graphs and and comparing them via scatter plot  

```python
sns.jointplot(data=tips,x='total_bill',y='tip')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='reg')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='kde')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hex')
```

```python
data=[2,3,4,5,1,2,6,7,9,10]
sns.displot(data,bins=4)
plt.show()
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
fig = px.scatter(iris, x = "sepal_length", color='species')
fig.show()
```

```python
fig = px.scatter_3d(iris, x = "sepal_length", y = "sepal_width", z = "petal_length", color = "species")
fig.show()
```

```python
fig = px.bar(iris, x = "sepal_length", y = "sepal_width", color = "species")
fig.show()
```

```python
fig = px.box(iris, x = "species", y = "sepal_width")
fig.show()
```
