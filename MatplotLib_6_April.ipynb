```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,60]
y=[10,20,15,18,16,30]
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y,ls='--',c='c',lw=5,marker='o',ms=20,mec='red',mfc='orange')
plt.show()
```

```python
#marker | Line |color

plt.plot(y,'^-b')
plt.show()
```

```python
plt.figure(figsize=(7,10))
plt.plot(x,y,marker='o')
plt.xlabel("time - >")
plt.ylabel("Temp")
plt.title("pollution temp graph")
plt.grid()
plt.show()
```
